```python
# Import our dependencies
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
import pandas as pd
import tensorflow as tf

#  Import and read the charity_data.csv.
import pandas as pd 
application_df = pd.read_csv("/content/drive/MyDrive/DataClassNotebooks/Module_21_HW/deep-learning-challenge/Code/charity_data.csv")
application_df.head()
```

```python
# Drop the non-beneficial ID columns, 'EIN' and 'NAME'.
df_reduced = application_df.drop(columns = ['EIN', 'NAME', 'ORGANIZATION', 'ASK_AMT'])

df_reduced
```

```python
# Determine the number of unique values in each column.
unique_counts = df_reduced.nunique(axis=0)
  
print("Unique values in each column :\n",
      unique_counts)
```

```python
# Look at APPLICATION_TYPE value counts for binning
df_reduced['APPLICATION_TYPE'].value_counts()
```

```python
# Choose a cutoff value and create a list of application types to be replaced
# use the variable name `application_types_to_replace`
application_types_to_replace = ['T25', 'T14', 'T29', 'T15', 'T17', 'T2', 'T12', 'T13', 'T9']

# Replace in dataframe
for app in application_types_to_replace:
    df_reduced['APPLICATION_TYPE'] = df_reduced['APPLICATION_TYPE'].replace(app,"Other")

# Check to make sure binning was successful
df_reduced['APPLICATION_TYPE'].value_counts()
```

```python
# Look at CLASSIFICATION value counts for binning
df_reduced['CLASSIFICATION'].value_counts()
```

```python
# You may find it helpful to look at CLASSIFICATION value counts >1

class_counts = df_reduced['CLASSIFICATION'].value_counts()

df_2 = pd.DataFrame(class_counts)
df_2.reset_index()
df_2


df_3 = df_2.loc[df_2['CLASSIFICATION'] < 50]
df_3

classifications_to_replace = df_3.index.values.tolist()

classifications_to_replace
```

```python

# Replace in dataframe
for cls in classifications_to_replace:
    df_reduced['CLASSIFICATION'] = df_reduced['CLASSIFICATION'].replace(cls,"Other")
    
# Check to make sure binning was successful
df_reduced['CLASSIFICATION'].value_counts()
```

```python
df_reduced
```

```python
# Convert categorical data to numeric with `pd.get_dummies`
dummies_df = pd.get_dummies(data = df_reduced, columns = ['APPLICATION_TYPE', 'CLASSIFICATION', 'INCOME_AMT', 'AFFILIATION', 'USE_CASE', 'SPECIAL_CONSIDERATIONS'])
```

```python
dummies_df
```

```python
# Split our preprocessed data into our features and target arrays
X = dummies_df.drop(columns='IS_SUCCESSFUL')

y = dummies_df['IS_SUCCESSFUL']


# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

## Compile, Train and Evaluate the Model

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.

nn_model = tf.keras.models.Sequential()




# First hidden layer
nn_model.add(tf.keras.layers.Dense(units=16, activation="relu", input_dim=48))

# Second hidden layer
nn_model.add(tf.keras.layers.Dense(units=16, activation="relu"))

# Third hidden layer

nn_model.add(tf.keras.layers.Dense(units=16, activation="relu"))

# Fourth hidden layer

nn_model.add(tf.keras.layers.Dense(units=16, activation="relu"))

# Fifth hidden layer

nn_model.add(tf.keras.layers.Dense(units=16, activation="relu"))

# Output layer
nn_model.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))

# Check the structure of the model
nn_model.summary()
```

```python
# Compile the model
nn_model.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn_model.fit(X_train_scaled, y_train, epochs=50)
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn_model.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
